```python
## MVAR & Voltage
import matplotlib.pyplot as plt

# Data for plotting
capacitor_MVAR = [220, 180, 140, 100, 50, 20]
percentage_increase = [38.40, 30.47, 23.04, 16.05, 7.83, 3.09]

# Plotting the relation between capacitor MVAR and percentage increase
plt.figure(figsize=(10, 6))
plt.plot(capacitor_MVAR, percentage_increase, marker='o', color='b', linestyle='-', linewidth=2, markersize=8)

# Adding labels and title
plt.xlabel('Capacitor Size (MVAR)', fontsize=12)
plt.ylabel('% Increase in Voltage', fontsize=12)
plt.title('Relation between Capacitor MVAR and % Increase in Voltage', fontsize=14)
plt.grid(True)

# Mark the data points
for i, txt in enumerate(percentage_increase):
    plt.annotate(f'{txt}%', (capacitor_MVAR[i], percentage_increase[i]), textcoords="offset points", xytext=(0, 10), ha='center')

# Display the graph
plt.show()
```

```python
##Bus Voltage vs Bus Angle () Graph
import matplotlib.pyplot as plt
import numpy as np

# Data for Bus No., Voltage (p.u.), and Angle ()
bus_no = np.array([1, 2, 3, 4, 5])
voltage_pu = np.array([1.06, 1.48, 1.023, 1.023, 1.018])
angle_deg = np.array([0, -2.75, -5.21, -5.49, -6.17])

# Plotting
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(bus_no, voltage_pu, 'bo-', label="Voltage (p.u.)")
plt.scatter(bus_no, voltage_pu, color='b')
plt.xlabel('Bus No.')
plt.ylabel('Voltage (p.u.)')
plt.title('Bus Voltage vs Bus Angle')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(bus_no, angle_deg, 'ro-', label="Angle ()")
plt.scatter(bus_no, angle_deg, color='r')
plt.xlabel('Bus No.')
plt.ylabel('Angle ()')
plt.grid(True)

plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for MATLAB vs ETAP comparison
bus_no = np.array([1, 2, 3, 4, 5])
matlab_values = {
    "voltage": np.array([1.06, 1.047, 1.022, 1.022, 1.017]),
    "angle": np.array([0, -2.752, -5.20, -5.48, -6.16]),
    "real_power": np.array([129.56, 20.013, -44.97, -39.97, -59.99]),
    "reactive_power": np.array([-7.06, 20.01, -14.99, -4.99, -9.99])
}
etap_values = {
    "voltage": np.array([1.06, 1.48, 1.023, 1.023, 1.018]),
    "angle": np.array([0, -2.75, -5.21, -5.49, -6.17]),
    "real_power": np.array([129.6, 20, 45, 40, 60]),
    "reactive_power": np.array([-7.05, 10, 15, 5, 10])
}

# Plotting MATLAB vs ETAP Voltage and Angle
plt.figure(figsize=(12, 8))

# Voltage Comparison
plt.subplot(2, 1, 1)
plt.plot(bus_no, matlab_values["voltage"], 'ro-', label="MATLAB Voltage")
plt.scatter(bus_no, matlab_values["voltage"], color='r')
plt.plot(bus_no, etap_values["voltage"], 'bo-', label="ETAP Voltage")
plt.scatter(bus_no, etap_values["voltage"], color='b')
plt.xlabel('Bus No.')
plt.ylabel('Voltage (p.u.)')
plt.title('MATLAB vs ETAP - Voltage')
plt.grid(True)
plt.legend()

# Mark MATLAB voltage data points
for i, txt in enumerate(matlab_values["voltage"]):
    plt.annotate(f'{txt:.3f}', (bus_no[i], matlab_values["voltage"][i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=9)

# Mark ETAP voltage data points
for i, txt in enumerate(etap_values["voltage"]):
    plt.annotate(f'{txt:.3f}', (bus_no[i], etap_values["voltage"][i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=9)

# Angle Comparison
plt.subplot(2, 1, 2)
plt.plot(bus_no, matlab_values["angle"], 'ro-', label="MATLAB Angle")
plt.scatter(bus_no, matlab_values["angle"], color='r')
plt.plot(bus_no, etap_values["angle"], 'bo-', label="ETAP Angle")
plt.scatter(bus_no, etap_values["angle"], color='b')
plt.xlabel('Bus No.')
plt.ylabel('Angle ()')
plt.title('MATLAB vs ETAP - Angle')
plt.grid(True)
plt.legend()

# Mark MATLAB angle data points
for i, txt in enumerate(matlab_values["angle"]):
    plt.annotate(f'{txt:.3f}', (bus_no[i], matlab_values["angle"][i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=9)

# Mark ETAP angle data points
for i, txt in enumerate(etap_values["angle"]):
    plt.annotate(f'{txt:.3f}', (bus_no[i], etap_values["angle"][i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for Voltage and Current Before and After Faults
bus_no = np.array([1, 2, 3, 4, 5])
before_fault_voltage = np.array([10.6, 14.8, 10.23, 10.23, 10.18])
lg_fault_voltage = np.array([10.6, 5.67, 10, 10, 6.08])
ll_fault_voltage = np.array([10.6, 5.12, 10, 10, 5.07])
llg_fault_voltage = np.array([10.6, 5.45, 10, 10, 5.98])

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(bus_no, before_fault_voltage, 'bo-', label='Before Fault Voltage')
plt.scatter(bus_no, before_fault_voltage, color='b')
plt.plot(bus_no, lg_fault_voltage, 'ro-', label='L-G Fault Voltage')
plt.scatter(bus_no, lg_fault_voltage, color='r')
plt.plot(bus_no, ll_fault_voltage, 'go-', label='L-L Fault Voltage')
plt.scatter(bus_no, ll_fault_voltage, color='g')
plt.plot(bus_no, llg_fault_voltage, 'mo-', label='L-L-G Fault Voltage')
plt.scatter(bus_no, llg_fault_voltage, color='m')

# Mark the data points for each fault condition
for i in range(len(bus_no)):
    plt.annotate(f'{before_fault_voltage[i]:.2f}', (bus_no[i], before_fault_voltage[i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=9, color='blue')
    plt.annotate(f'{lg_fault_voltage[i]:.2f}', (bus_no[i], lg_fault_voltage[i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=9, color='red')
    plt.annotate(f'{ll_fault_voltage[i]:.2f}', (bus_no[i], ll_fault_voltage[i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=9, color='green')
    plt.annotate(f'{llg_fault_voltage[i]:.2f}', (bus_no[i], llg_fault_voltage[i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=9, color='magenta')

plt.xlabel('Bus No.')
plt.ylabel('Voltage (kV)')
plt.title('Voltage Comparison Before and After Fault Initiation')
plt.grid(True)
plt.legend()
plt.show()
```
